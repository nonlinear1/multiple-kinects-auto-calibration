```python
from os import listdir
import numpy as np
import matplotlib.pylab as plt 
from __future__ import division
%matplotlib inline  

def get_path(cam, plane):
    return "../hist/cam" + str(cam) + "/plane" + str(plane) + "/"

def list_files(cam, plane):
    path = get_path(cam, plane)
    return [path + file for file in listdir(path)]


def read_array(file_reader, n_lines):
    return [[int(i) for i in file_reader.readline().split()] for _ in range(n_lines)]


def read_file(path):
    with open(path, 'r') as f:
        f.readline()
        hist = read_array(f, 256)
    return np.array(hist)

def plot_hist(hist, cam=-1, plane=-1):
    plot_colors = ['r', 'g', 'b']
    plt.figure(figsize=[20,10])
    for i in range(3):
        color = hist[:,i]
        plt.bar(np.linspace(0, 255, len(color)), color, width=1, color=plot_colors[i], alpha=0.5)
    plt.legend(["red channel", "green channel", "blue channel"])
    if cam != -1 and plane != -1:
        plt.title("Color histogram of a cam" + str(cam) + " plane" + str(plane))
    else:
        plt.title("Color histogram")
        
def average_hist(list_hist):
    n = len(list_hist)
    return sum(list_hist)/n
```

```python
_cams = [1, 2, 3]
_planes = [1, 2, 3]

for plane in _planes:
    print("plane" + str(plane))
    for cam in _cams:
        print("\tcam" + str(cam))
        files = list_files(cam, plane)
        hist = average_hist([read_file(files[i]) for i in range(len(files))])
        plot_hist(hist, cam, plane)
```

```python
from scipy.spatial import distance as dist

SCIPY_METHODS = (
    ("Euclidean", dist.euclidean),
    ("Manhattan", dist.cityblock),
    ("Chebyshev", dist.chebyshev))

test_cam = 1
test_plane = 1
test_id = 0

test_file = list_files(test_cam, test_plane)[test_id]
test_hist = read_file(test_file)

result = []

print("Computing distance to " + test_file)

for (methodName, method) in SCIPY_METHODS:
    
    print(methodName + " distance")
    
    for plane in _planes:
        for cam in _cams:
            print("\tcam" + str(cam) + " plane" + str(plane))
            files = list_files(cam, plane)
            
            for i in range(len(files)):
                hist = read_file(files[i])
                line = [plane, methodName]
                
                dist = method(hist[:,0], test_hist[:,0])
                print("\t\tR: " + str(dist))
                line.append(dist)
                
                dist = method(hist[:,1], test_hist[:,1])
                print("\t\tG: " + str(dist))
                line.append(dist)
                
                dist = method(hist[:,2], test_hist[:,2])
                print("\t\tB: " + str(dist))
                line.append(dist)
                
                dist = method(np.linalg.norm(hist, axis=1), np.linalg.norm(test_hist, axis=1))
                print("\t\tGray: " + str(dist) + "\n")
                line.append(dist)
                result.append(line)
```

```python
for (methodName, method) in SCIPY_METHODS:
    print(methodName + " distance")
    for plane in _planes:
        print("\tPlane " + str(plane) + ": " + str(np.mean([r[2:] for r in result if r[0] == plane and r[1] == methodName], axis=0)))
```

```python
from scipy.spatial import distance as dist

(methodName, method) = (("Manhattan", dist.cityblock))

test_cam = 1
test_plane = 1
test_id = 0

test_file = list_files(test_cam, test_plane)[test_id]
test_hist = read_file(test_file)

result = []

print("Computing distance to " + test_file)
    
print(methodName + " distance")

for plane in _planes:
    for cam in _cams:
        #print("\tcam" + str(cam) + " plane" + str(plane))
        files = list_files(cam, plane)

        for i in range(len(files)):
            hist = read_file(files[i])
            line = [plane]

            dist = method(hist[:,0], test_hist[:,0])
            #print("\t\tR: " + str(dist))
            line.append(dist)

            dist = method(hist[:,1], test_hist[:,1])
            #print("\t\tG: " + str(dist))
            line.append(dist)

            dist = method(hist[:,2], test_hist[:,2])
            #print("\t\tB: " + str(dist))
            line.append(dist)

            dist = method(np.linalg.norm(hist, axis=1), np.linalg.norm(test_hist, axis=1))
            #print("\t\tGray: " + str(dist) + "\n")
            line.append(dist)
            result.append(line)
            
color = ["", "r", "g", "b"]
plt.figure()
for plane in _planes:
    plt.scatter([r[1] for r in result if r[0] == plane], [r[2] for r in result if r[0] == plane], [r[3] for r in result if r[0] == plane], marker="x", c=color[plane])    
plt.legend(["plane" + str(p) for p in _planes])
plt.xlabel("R")
plt.ylabel("G")
plt.zlabel("B")
```
