```python
import numpy as np
import matplotlib.pylab as plt
from scipy.stats import kurtosis
%matplotlib inline

def read_log(file):
    with open(file) as f:
        data = [[float(n) for n in l.split()] for l in f]
    return np.array(data)
```

```python
class LogReader:
    def __init__(self, file, name=''):
        if name == '':
            name = file.split('/')[-1].split('.')[0]
        self.name = name
        with open(file) as f:
            print("Loading data...")
            data = []
            for i, l in enumerate(f):
                if i%1000 == 0 and i!= 0:
                    print(i)
                data.append([float(x) for x in l.split()])
            self.data = np.array(data)
        self.v_t = self.data[:,:3]
        self.t = np.linalg.norm(self.v_t, axis=1)
        self.a_t = self.data[:,3:6]
        self.th = self.data[:,6] * 180 / 3.14159
        
    def cut_outliers(self, vec):
        return np.array([x for x in vec if (x >= np.percentile(vec, 1) and x <= np.percentile(vec, 99)) ])
    
    def stats(self, vec):
        vec = self.cut_outliers(vec)
        return np.mean(vec), np.std(vec), kurtosis(vec)
    
    def plot_hist(self, vec):
        vec = self.cut_outliers(vec)
        plt.hist(vec, 100)
```

```python
LR = LogReader('cam1_cam2.log')
print(LR.stats(LR.t))
LR.plot_hist(LR.t)
```

```python
print(LR.stats(LR.th))
LR.plot_hist(LR.th)
```
